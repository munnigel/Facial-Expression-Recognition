```python
import os
import matplotlib.pyplot as plt

# Set up the paths to the training and validation image directories
train_dir = 'datasets/train'
val_dir = 'datasets/validation'

# Define a function to count the number of images in each emotion category
def count_images(directory):
    emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
    counts = []
    for emotion in emotions:
        emotion_dir = os.path.join('../', directory, emotion.replace("\\\\", "/"))
        count = len(os.listdir(emotion_dir))
        counts.append(count)
    return counts

# Get the counts of images for each emotion category in the training and validation data
train_counts = count_images(train_dir)
val_counts = count_images(val_dir)

# Define the emotions as the x-axis labels for the bar graphs
emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']

# Define the colors for each emotion category
colors = ['#ff0000', '#8b0000', '#ff4500', '#ffd700', '#c0c0c0', '#708090', '#00ffff']

# Plot the bar graphs for the training and validation data
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].bar(emotions, train_counts, color=colors)
ax[0].set_title('Training Data')
ax[0].set_xlabel('Emotions')
ax[0].set_ylabel('Count')

ax[1].bar(emotions, val_counts, color=colors)
ax[1].set_title('Validation Data')
ax[1].set_xlabel('Emotions')
ax[1].set_ylabel('Count')

plt.show()
```
