```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("../datasets/ckextended.csv")
training_counts = data[data['Usage'] == 'Training']['emotion'].value_counts().sort_index()
public_test_counts = data[data['Usage'] == 'PublicTest']['emotion'].value_counts().sort_index()
emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart for Training data
ax1.bar(emotions, training_counts)
ax1.set_title('Training Data')
ax1.set_xlabel('Emotions')
ax1.set_ylabel('Count')

# Bar chart for PublicTest data
ax2.bar(emotions, public_test_counts)
ax2.set_title('Test Data')
ax2.set_xlabel('Emotions')
ax2.set_ylabel('Count')

# Display the bar charts
plt.show()
```
