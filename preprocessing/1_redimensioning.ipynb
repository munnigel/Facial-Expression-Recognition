```python
from PIL import Image
from IPython.display import display



# Open the original image
img = Image.open('../datasets/train/disgust/9913.jpg')

# Get the input image pixel size
input_size = img.size
print('Input image pixel size:', input_size)


# Resize the image
if input_size[0] < 256 or input_size[1] < 256:
    output_size = (256, 256 * input_size[1] // input_size[0])

img_resized = img.resize(output_size)

# Get the output image pixel size
print('Output image pixel size:', img_resized.size)

# Display both images
display(img)
display(img_resized)

# Save the resized image
img_resized.save('533_resized.jpg')
```
